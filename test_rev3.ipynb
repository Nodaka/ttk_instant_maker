```python
def read_excel_onedrive(
    file_path,
    sheet_name,
    index_col,
    ):
    copied_file_path = os.path.splitext(file_path)[0] + '_instant'+ os.path.splitext(file_path)[1]
    command_str = 'echo F | xcopy ' + '"' + file_path + '" "' + copied_file_path + '"' +' /Y'
    proc = subprocess.run(command_str, shell=True)
    return_df = pd.read_excel(copied_file_path, sheet_name=sheet_name, index_col=index_col)
    
    delete_str = 'del ' + '"' + copied_file_path + '"'
    proc = subprocess.run(delete_str, shell=True)
    
    return return_df
```

```python
def widget_place(self, widget_ds):
    widget_ds = widget_ds.replace([float('nan')], [None])

    if widget_ds['type'] == 'ttk.Frame':

        options_list = ["cursor", "relief"]
        getattr_options_list = ["master"]

        self.ttk_style = ttk.Style()
        self.ttk_style.configure(
            widget_ds.name + '.TFrame', 
            #font=(widget_ds['font'], widget_ds['font_size']), 
            #foreground = widget_ds['foreground'],
            background = widget_ds['background'],
            )

        return_widget = ttk.Frame(
            style=widget_ds.name + '.TFrame',
            )

        for getattr_option in getattr_options_list:
            try:
                return_widget[getattr_option] = getattr(self, widget_ds[getattr_option])
            except:
                pass
        for option in options_list:
            try:
                return_widget[option] = widget_ds[option]
            except:
                pass

        return_widget.place(
            x=widget_ds['x'], 
            y=widget_ds['y'], 
            width=widget_ds['width'], 
            height=widget_ds['height'])
        
        return return_widget
    elif widget_ds['type'] == 'ttk.Label':

        options_list = ['text', 'relief', 'cursor', "anchor"]
        getattr_options_list = ["master", "textvariable"]

        self.ttk_style.configure(
            widget_ds.name + '.TLabel', 
            font=(widget_ds['font'], widget_ds['font_size']), 
            foreground = widget_ds['foreground'],
            background = widget_ds['background'],
            )
        
        return_widget = ttk.Label(
            style=widget_ds.name + '.TLabel',
            )

        for getattr_option in getattr_options_list:
            try:
                return_widget[getattr_option] = getattr(self, widget_ds[getattr_option])
            except:
                pass
        for option in options_list:
            try:
                return_widget[option] = widget_ds[option]
            except:
                pass
            
        return_widget.place(
            x=widget_ds['x'],
            y=widget_ds['y'],
            width=widget_ds['width'],
            height=widget_ds['height'],
            )
        return return_widget
    elif widget_ds['type'] == 'ttk.Button':
        getattr_options_list = ["master", 'command', 'textvariable']
        options_list = ['text', "cursor", "anchor", "state"]

        self.ttk_style.configure(
            widget_ds.name + '.TButton', 
            font=(widget_ds['font'], widget_ds['font_size']), 
            #bordercolor = "Green",
            #highlightbackground = "Green",
            #highlightcolor = "Green",
            #foreground = widget_ds['foreground'],
            #background = widget_ds['background'],
            )
        self.ttk_style.map(
            widget_ds.name + '.TButton', 
            background = [("active", "Red"), ("disabled", "Black")],
            foreground = [("active", "Red"), ("disabled", "Black")],
        )
        return_widget = ttk.Button(
            style=widget_ds.name + '.TButton',
            )

        for getattr_option in getattr_options_list:
            try:
                return_widget[getattr_option] = getattr(self, widget_ds[getattr_option])
            except:
                pass
        for option in options_list:
            try:
                return_widget[option] = widget_ds[option]
            except:
                pass            
        return_widget.place(
            x=widget_ds['x'],
            y=widget_ds['y'],
            width=widget_ds['width'],
            height=widget_ds['height'],
            )
        return return_widget
    
    elif widget_ds['type'] == 'ttk.Entry':
        getattr_options_list = []
        variable_options_list = ['textvariable', ]
        options_list = ['text']

        self.ttk_style.configure(
            widget_ds.name + '.TEntry', 
            font=(widget_ds['font'], widget_ds['font_size']), 
            anchor = widget_ds['anchor'],
            foreground = widget_ds['foreground'],
            background = widget_ds['background'],
            )
        return_widget = ttk.Entry(
            master=self.widget_dict[widget_ds['master']],
            style=widget_ds.name + '.TEntry',
            cursor=widget_ds['cursor'],
            )

        for option in getattr_options_list:
            try:
                return_widget[option] = getattr(self, option + '_' + widget_ds.name)
            except:
                pass
        for option in variable_options_list:
            try:
                return_widget[option] = self.__dict__[option + '_' + widget_ds.name]
            except:
                pass
        for option in options_list:
            return_widget[option] = widget_ds[option]
            
        return_widget.place(
            x=widget_ds['x'],
            y=widget_ds['y'],
            width=widget_ds['width'],
            height=widget_ds['height'],
            )
        return return_widget
    elif widget_ds['type'] == 'ttk.Checkbutton':
        getattr_options_list = ['command']
        variable_options_list = ['variable', ]
        options_list = ['text', 'state']

        self.ttk_style.configure(
            widget_ds.name + '.TCheckbutton', 
            font=(widget_ds['font'], widget_ds['font_size']), 
            anchor = widget_ds['anchor'],
            foreground = widget_ds['foreground'],
            background = widget_ds['background'],
            )
        return_widget = ttk.Checkbutton(
            master=self.widget_dict[widget_ds['master']],
            style=widget_ds.name + '.TCheckbutton',
            cursor=widget_ds['cursor'],
            )

        for option in getattr_options_list:
            try:
                return_widget[option] = getattr(self, option + '_' + widget_ds.name)
            except:
                pass
        for option in variable_options_list:
            try:
                return_widget[option] = self.__dict__[option + '_' + widget_ds.name]
            except:
                pass
        for option in options_list:
            return_widget[option] = widget_ds[option]
            
        return_widget.place(
            x=widget_ds['x'],
            y=widget_ds['y'],
            width=widget_ds['width'],
            height=widget_ds['height'],
            )
        print(return_widget['state'])
        return return_widget
    elif widget_ds['type'] == 'ttk.Radiobutton':
        getattr_options_list = []
        variable_options_list = []
        variable_sel_options_list = ['variable']
        options_list = ['value', 'relief', 'text']
        
        self.ttk_style.configure(
            widget_ds.name + '.TRadiobutton', 
            font=(widget_ds['font'], widget_ds['font_size']), 
            anchor = widget_ds['anchor'],
            foreground = widget_ds['foreground'],
            background = widget_ds['background'],
            )
        return_widget = ttk.Radiobutton(
            master=self.widget_dict[widget_ds['master']],
            style=widget_ds.name + '.TRadiobutton',
            cursor=widget_ds['cursor'],
            )

        for option in getattr_options_list:
            try:
                return_widget[option] = getattr(self, option + '_' + widget_ds.name)
            except:
                pass
        for option in variable_options_list:
            try:
                return_widget[option] = self.__dict__[option + '_' + widget_ds.name]
            except:
                pass
        for option in variable_sel_options_list:
            try:
                return_widget[option] = self.__dict__[widget_ds[option]]
            except:
                pass
        for option in options_list:
            return_widget[option] = widget_ds[option]
            
        return_widget.place(
            x=widget_ds['x'],
            y=widget_ds['y'],
            width=widget_ds['width'],
            height=widget_ds['height'],
            )
        return return_widget
    elif widget_ds['type'] == 'tk.Listbox':
        getattr_options_list = []
        variable_options_list = ['listvariable']
        variable_sel_options_list = []
        options_list = ['text', 'value', 'foreground', 'background', 'cursor', 'selectmode']
        
        return_widget = tk.Listbox(
            master=self.widget_dict[widget_ds['master']],
            )

        for option in getattr_options_list:
            try:
                return_widget[option] = getattr(self, option + '_' + widget_ds.name)
            except:
                pass
        for option in variable_options_list:
            try:
                return_widget[option] = self.__dict__[option + '_' + widget_ds.name]
            except:
                pass
        for option in variable_sel_options_list:
            try:
                return_widget[option] = self.__dict__[widget_ds[option]]
            except:
                pass
        for option in options_list:
            return_widget[option] = widget_ds[option]
            
        return_widget.place(
            x=widget_ds['x'],
            y=widget_ds['y'],
            width=widget_ds['width'],
            height=widget_ds['height'],
            )
        return return_widget
    elif widget_ds['type'] == 'ttk.Scale':
        getattr_options_list = []
        variable_options_list = ['variable']
        variable_sel_options_list = []
        options_list = ['orient' ,'value', 'from_', 'to']
        
        if widget_ds['orient'] == 'vertical':
            self.ttk_style.configure(
                widget_ds.name + '.Vertical.TScale', 
                font=(widget_ds['font'], widget_ds['font_size']), 
                anchor = widget_ds['anchor'],
                foreground = widget_ds['foreground'],
                background = widget_ds['background'],
                )

            return_widget = ttk.Scale(
                master=self.widget_dict[widget_ds['master']],
                cursor=widget_ds['cursor'],
                style=widget_ds.name + '.Vertical.TScale',
                )
        else:
            self.ttk_style.configure(
                widget_ds.name + '.Horizontal.TScale', 
                font=(widget_ds['font'], widget_ds['font_size']), 
                anchor = widget_ds['anchor'],
                foreground = widget_ds['foreground'],
                background = widget_ds['background'],
                )

            return_widget = ttk.Scale(
                master=self.widget_dict[widget_ds['master']],
                cursor=widget_ds['cursor'],
                style=widget_ds.name + '.Horizontal.TScale',
                )

        for option in getattr_options_list:
            try:
                return_widget[option] = getattr(self, option + '_' + widget_ds.name)
            except:
                pass
        for option in variable_options_list:
            try:
                return_widget[option] = self.__dict__[option + '_' + widget_ds.name]
            except:
                pass
        for option in variable_sel_options_list:
            try:
                return_widget[option] = self.__dict__[widget_ds[option]]
            except:
                pass
        for option in options_list:
            return_widget[option] = widget_ds[option]
            
        return_widget.place(
            x=widget_ds['x'],
            y=widget_ds['y'],
            width=widget_ds['width'],
            height=widget_ds['height'],
            )
        return return_widget
    elif widget_ds['type'] == 'ttk.Combobox':
        getattr_options_list = ['postcommand']
        variable_options_list = ['values', 'textvariable']
        variable_sel_options_list = []
        options_list = []
        
        self.ttk_style.configure(
            widget_ds.name + '.TCombobox', 
            font=(widget_ds['font'], widget_ds['font_size']), 
            anchor = widget_ds['anchor'],
            foreground = widget_ds['foreground'],
            background = widget_ds['background'],
            )
        return_widget = ttk.Combobox(
            master=self.widget_dict[widget_ds['master']],
            style=widget_ds.name + '.TCombobox',
            cursor=widget_ds['cursor'],
            )

        for option in getattr_options_list:
            try:
                return_widget[option] = getattr(self, option + '_' + widget_ds.name)
            except:
                pass
        for option in variable_options_list:
            try:
                return_widget[option] = self.__dict__[option + '_' + widget_ds.name]
            except:
                pass
        for option in variable_sel_options_list:
            try:
                return_widget[option] = self.__dict__[widget_ds[option]]
            except:
                pass
        for option in options_list:
            return_widget[option] = widget_ds[option]
            
        return_widget.place(
            x=widget_ds['x'],
            y=widget_ds['y'],
            width=widget_ds['width'],
            height=widget_ds['height'],
            )
        return return_widget
    elif widget_ds['type'] == 'ttk.Treeview':
        getattr_options_list = []
        variable_options_list = []
        variable_sel_options_list = []
        options_list = []
        
        self.ttk_style.configure(
            widget_ds.name + '.Treeview', 
            font=(widget_ds['font'], widget_ds['font_size']), 
            anchor = widget_ds['anchor'],
            foreground = widget_ds['foreground'],
            background = widget_ds['background'],
            )
        return_widget = ttk.Treeview(
            master=self.widget_dict[widget_ds['master']],
            style=widget_ds.name + '.Treeview',
            cursor=widget_ds['cursor'],
            )

        print(return_widget['selectmode'])

        return_widget['selectmode'] = 'browse'
        print(widget_ds['selectmode'])
        return_widget['selectmode'] = str(widget_ds['selectmode'])

        for option in getattr_options_list:
            try:
                return_widget[option] = getattr(self, option + '_' + widget_ds.name)
            except:
                pass
        for option in variable_options_list:
            try:
                return_widget[option] = self.__dict__[option + '_' + widget_ds.name]
            except:
                pass
        for option in variable_sel_options_list:
            try:
                return_widget[option] = self.__dict__[widget_ds[option]]
            except:
                pass
        for option in options_list:
            return_widget[option] = widget_ds[option]
            
        return_widget.place(
            x=widget_ds['x'],
            y=widget_ds['y'],
            width=widget_ds['width'],
            height=widget_ds['height'],
            )
        return return_widget

    elif widget_ds['type'] == 'FigureCanvasTkAgg':
        return_canvas =  FigureCanvasTkAgg(
            self.__dict__[widget_ds['command']],
            self.widget_dict[widget_ds['frame']],
            )

        return_widget = return_canvas.get_tk_widget()
        return_widget['cursor'] = widget_ds['cursor']
        return_widget.place(
            x=widget_ds['x'],
            y=widget_ds['y'],
            width=widget_ds['width'],
            height=widget_ds['height'],
            )
        return [return_canvas, return_widget]
    elif widget_ds['type'] == 'ttk.Labelframe':
        self.ttk_style = ttk.Style()
        self.ttk_style.configure(
            widget_ds.name + '.TLabelframe',
            background = widget_ds['background'],)
        
        return_widget = ttk.Labelframe(
            self.widget_dict[widget_ds['frame']],
            text=widget_ds['text'],
            style=widget_ds.name + '.TLabelframe',
            cursor=widget_ds['cursor'],
            relief=widget_ds['relief'],
            labelanchor=widget_ds['anchor']
            )
        return_widget.place(
            x=widget_ds['x'],
            y=widget_ds['y'],
            width=widget_ds['width'],
            height=widget_ds['height'],
            )
        return return_widget
    elif widget_ds['type'] == 'ttk.Menubutton':
        self.ttk_style = ttk.Style()
        self.ttk_style.configure(
            widget_ds.name + '.TMenubutton',
            background = widget_ds['background'],)
        
        return_widget = ttk.Menubutton(
            self.widget_dict[widget_ds['frame']],
            text=widget_ds['text'],
            style=widget_ds.name + '.TMenubutton',
            menu=self.__dict__[widget_ds.name],
            compound=widget_ds['anchor'],
            )
        return_widget.place(
            x=widget_ds['x'],
            y=widget_ds['y'],
            width=widget_ds['width'],
            height=widget_ds['height'],
            )
        return return_widget
```
