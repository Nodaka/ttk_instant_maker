```python
def widget_place(self, widget_ds):
    widget_ds = widget_ds.replace([float('nan')], [None])
    if widget_ds['type'] == 'ttk.Button':
        self.ttk_style = ttk.Style()
        self.ttk_style.configure(
            widget_ds.name + '.TButton', 
            font=(widget_ds['font'], widget_ds['font_size']), 
            anchor = widget_ds['anchor'],
            foreground = widget_ds['foreground'],
            background = widget_ds['background'],
            )
        if widget_ds['text'] != None:
            return_widget = ttk.Button(
                self.widget_dict[widget_ds['frame']],
                command = getattr(self, widget_ds['command']),
                text=widget_ds['text'],
                style=widget_ds.name + '.TButton',
                cursor=widget_ds['cursor'],
                )
        else:
            return_widget = ttk.Button(
                self.widget_dict[widget_ds['frame']],
                command = getattr(self, widget_ds['command']),
                textvariable=self.__dict__[widget_ds.name],
                style=widget_ds.name + '.TButton',
                cursor=widget_ds['cursor'],
                )

        return_widget.place(
            x=widget_ds['x'],
            y=widget_ds['y'],
            width=widget_ds['width'],
            height=widget_ds['height'],
            )
        return return_widget
    
    elif widget_ds['type'] == 'ttk.Entry':
        self.ttk_style = ttk.Style()
        self.ttk_style.configure(
            widget_ds.name + '.TEntry', 
            font=(widget_ds['font'], widget_ds['font_size']), 
            anchor = widget_ds['anchor'],
            foreground = widget_ds['foreground'],
            background = widget_ds['background'],
            )
        return_widget = ttk.Entry(
            self.widget_dict[widget_ds['frame']],
            textvariable=self.__dict__[widget_ds.name],
            style=widget_ds.name + '.TEntry',
            cursor=widget_ds['cursor'],
            )
        return_widget.place(
            x=widget_ds['x'],
            y=widget_ds['y'],
            width=widget_ds['width'],
            height=widget_ds['height'],
            )
        return return_widget
    elif widget_ds['type'] == 'ttk.Label':
        self.ttk_style = ttk.Style()
        self.ttk_style.configure(
            widget_ds.name + '.TLabel', 
            font=(widget_ds['font'], widget_ds['font_size']), 
            anchor = widget_ds['anchor'],
            foreground = widget_ds['foreground'],
            background = widget_ds['background'],
            )
        
        if widget_ds['text'] != None:
            return_widget = ttk.Label(
                self.widget_dict[widget_ds['frame']],
                text=widget_ds['text'],
                style=widget_ds.name + '.TLabel',
                cursor=widget_ds['cursor'],
                )
        else:
            return_widget = ttk.Label(
                self.widget_dict[widget_ds['frame']],
                textvariable=self.__dict__[widget_ds.name],
                style=widget_ds.name + '.TLabel',
                cursor=widget_ds['cursor'],
                )

        return_widget.place(
            x=widget_ds['x'],
            y=widget_ds['y'],
            width=widget_ds['width'],
            height=widget_ds['height'],
            )
        return return_widget
    
    elif widget_ds['type'] == 'ttk.Frame':
        self.ttk_style = ttk.Style()
        self.ttk_style.configure(
            widget_ds.name + '.TFrame',
            background = widget_ds['background'],)

        if widget_ds['frame'] == None:
            return_widget = ttk.Frame(
                self.master,
                style=widget_ds.name + '.TFrame',
                cursor=widget_ds['cursor'],
                )
        else:
            return_widget = ttk.Frame(
                self.widget_dict[widget_ds['frame']],
                style='instant.TFrame',
                cursor=widget_ds['cursor'],
                )
        return_widget.place(
            x=widget_ds['x'],
            y=widget_ds['y'],
            width=widget_ds['width'],
            height=widget_ds['height'],
            )
        return return_widget
    elif widget_ds['type'] == 'FigureCanvasTkAgg':
        return_canvas =  FigureCanvasTkAgg(
            self.__dict__[widget_ds['command']],
            self.widget_dict[widget_ds['frame']],
            )

        return_widget = return_canvas.get_tk_widget()
        return_widget['cursor'] = widget_ds['cursor']
        return_widget.place(
            x=widget_ds['x'],
            y=widget_ds['y'],
            width=widget_ds['width'],
            height=widget_ds['height'],
            )
        return [return_canvas, return_widget]
    elif widget_ds['type'] == 'ttk.Checkbutton':
        self.ttk_style = ttk.Style()
        self.ttk_style.configure(
            widget_ds.name + '.TCheckbutton',
            background = widget_ds['background'],)
        
        return_widget = ttk.Checkbutton(
            self.widget_dict[widget_ds['frame']],
            command = getattr(self, widget_ds['command']),
            variable=self.__dict__[widget_ds.name],
            style=widget_ds.name + '.TCheckbutton',
            cursor=widget_ds['cursor'],
            )
        return_widget.place(
            x=widget_ds['x'],
            y=widget_ds['y'],
            width=widget_ds['width'],
            height=widget_ds['height'],
            )
        return return_widget
    elif widget_ds['type'] == 'ttk.Labelframe':
        self.ttk_style = ttk.Style()
        self.ttk_style.configure(
            widget_ds.name + '.TLabelframe',
            background = widget_ds['background'],)
        
        return_widget = ttk.Labelframe(
            self.widget_dict[widget_ds['frame']],
            text=widget_ds['text'],
            style=widget_ds.name + '.TLabelframe',
            cursor=widget_ds['cursor'],
            relief=widget_ds['relief'],
            labelanchor=widget_ds['anchor']
            )
        return_widget.place(
            x=widget_ds['x'],
            y=widget_ds['y'],
            width=widget_ds['width'],
            height=widget_ds['height'],
            )
        return return_widget
    elif widget_ds['type'] == 'ttk.Menubutton':
        self.ttk_style = ttk.Style()
        self.ttk_style.configure(
            widget_ds.name + '.TMenubutton',
            background = widget_ds['background'],)
        
        return_widget = ttk.Menubutton(
            self.widget_dict[widget_ds['frame']],
            text=widget_ds['text'],
            style=widget_ds.name + '.TMenubutton',
            )
        return_widget.place(
            x=widget_ds['x'],
            y=widget_ds['y'],
            width=widget_ds['width'],
            height=widget_ds['height'],
            )
        return return_widget
```

```python
mpl.use('Agg')
class manual_annotation(tk.Frame):
    def __init__(self, master=None):
        super().__init__(master)
        self.master = master
        self.master.title('test')
       
        self.widget_df = pd.read_excel('widget.xlsx', sheet_name='widget', index_col=0)
        self.widget_dict = {}

        self.folder_path_entry = tk.StringVar()
        self.ref_button = tk.StringVar(value='ref')
        self.test_label = tk.StringVar(value='test')
        self.next_button = tk.StringVar(value='next')
        self.back_button = tk.StringVar(value='back')

        self.instant_checkbutton = tk.BooleanVar(value=False)
        self.instant_labelframe = tk.StringVar(value='labelframe')

        self.figure_001 = plt.figure(facecolor="azure")
        self.ax_001 = self.figure_001.add_axes([0, 0, 1, 1])

        self.img_file_path_list = []
        self.img_array_list = []
        self.img_index = 0

        self.trim_flag = False
        self.trim_sta_x_posi = 0
        self.trim_sta_y_posi = 0
        self.trim_end_x_posi = 0
        self.trim_end_y_posi = 0
        self.trim_patch = None

        for col in self.widget_df.columns:
            self.widget_dict[col] = widget_place(self, self.widget_df[col])


        self.master.bind("<ButtonPress-1>", self.print_cord)
        self.master.bind("<Shift-ButtonPress-1>", self.trimming_start)
        self.master.bind("<Shift-B1-Motion>", self.trimming)
        self.master.bind("<Shift-ButtonRelease-1>", self.trimming_end)
    def action_instant_checkbutton(self):
        return None
    
    def trimming_start(self, event):
        if (len(self.img_file_path_list)>0)*(event.widget == self.widget_dict['figure_canvas'][1]):
            self.trim_flag = True
            self.trim_sta_x_posi = event.x
            self.trim_sta_y_posi = event.y
            self.trim_patch = patches.Rectangle(
                (self.trim_sta_x_posi, self.trim_sta_y_posi),
                width = 0,
                height = 0,
                )
            self.ax_001.add_patch(self.trim_patch)
    def trimming(self, event):
        if self.trim_flag:
            self.trim_end_x_posi = event.x
            self.trim_end_y_posi = event.y
            self.trim_patch.remove()

            self.trim_patch = patches.Rectangle(
                (min(self.trim_sta_x_posi, self.trim_end_x_posi), min(self.trim_sta_y_posi, self.trim_end_y_posi)),
                width = abs(self.trim_sta_x_posi - self.trim_end_x_posi),
                height = abs(self.trim_sta_y_posi - self.trim_end_y_posi),
                fill = False,
                ec = 'Magenta',
                )
            self.ax_001.add_patch(self.trim_patch)
            self.widget_dict['figure_canvas'][0].draw()
        
    def trimming_end(self, event):
        self.widget_dict['figure_canvas'][1]['cursor'] = 'arrow'
        self.trim_flag = False


    def print_cord(self, event):
        if (len(self.img_file_path_list)>0)*(event.widget == self.widget_dict['figure_canvas'][1]):
            self.ax_001.axhline(event.y)
            self.ax_001.axvline(event.x)
            self.widget_dict['figure_canvas'][0].draw()
    def action_ref_button(self):
        instant_folder_path = filedialog.askdirectory()
        self.folder_path_entry.set(instant_folder_path)
        self.test_thread = threading.Thread(target=self.thread_load_files)
        self.test_thread.start()
        return None
    def thread_load_files(self):
        instant_folder_path = self.folder_path_entry.get()
        file_path_list = []
        file_path_list = file_path_list + glob.glob(instant_folder_path + '//*.jpg')
        file_path_list = file_path_list + glob.glob(instant_folder_path + '//*.png')
        if len(file_path_list)==0:
            return None
        self.img_file_path_list = file_path_list.copy()
        self.img_array_list = []
        for file_path in file_path_list:
            instant_img_array = cv2.imread(file_path, 0)
            self.img_array_list.append(cv2.resize(instant_img_array, (int(self.widget_df['figure_canvas']['width']), int(self.widget_df['figure_canvas']['height']))))
        self.update_ax_001(0)
    def update_ax_001(self, img_index):
        if (len(self.img_file_path_list)>0)*(img_index>=0)*(img_index<len(self.img_file_path_list)):
            self.ax_001.cla()
            self.ax_001.imshow(self.img_array_list[img_index], 'gray')
            self.ax_001.set_ylim(self.img_array_list[img_index].shape[0], 0)
            self.ax_001.set_xlim(0, self.img_array_list[img_index].shape[1])
            self.widget_dict['figure_canvas'][0].draw()
            self.img_index = img_index
        
    def action_back_button(self):
        self.update_ax_001(self.img_index - 1)
        return None
    def action_next_button(self):
        self.update_ax_001(self.img_index + 1)
        return None

root = tk.Tk()
application = manual_annotation(root)
application.mainloop()
plt.close()
```
